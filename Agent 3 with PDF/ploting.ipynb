```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Results/results_v1.csv")
df2=pd.read_csv("Results/results_v2.csv")
df.head(2)
```

### Individually

```python
# ---------------------------
# AGENT V1
# ---------------------------

# ---------------------------
# 1. Answer Relevancy Comparison
# ---------------------------
plt.figure(figsize=(13, 4))
plt.plot(df["user_input"], df["answer_relevancy"], marker='o', label="Answer_relevancy",color="blue")
plt.title("Answer relevancy")
plt.xlabel("Question")
plt.ylabel("Accuracy")
plt.ylim(0, 1.1)
plt.xticks(rotation=15, ha='right', fontsize=12)  
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Images/v1/answer_relevancy.png", dpi=300)  
plt.show()

# ---------------------------
# 2. Context Precision Comparison
# ---------------------------
plt.figure(figsize=(13, 4))
plt.plot(df["user_input"], df["context_precision_no_ref"], marker='o', label="Context_precision",color="blue")
plt.title("Context precision")
plt.xlabel("Question")
plt.ylabel("Accuracy")
plt.ylim(0, 1.1)
plt.xticks(rotation=15, ha='right', fontsize=12)  
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Images/v1/context_precision.png", dpi=300)  
plt.show()

# ---------------------------
# 3. Faithfulness Comparison
# ---------------------------
plt.figure(figsize=(13, 4))
plt.plot(df2["user_input"], df["faithfulness"], marker='s', label="Faithfulness",color="blue")
plt.title("Faithfulness ")
plt.xlabel("Question")
plt.ylabel("Accuracy")
plt.ylim(0, 1.1)
plt.xticks(rotation=15, ha='right', fontsize=12)  
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Images/v1/faithfulness.png", dpi=300)  
plt.show()


# ---------------------------
# AGENT V2
# ---------------------------

# ---------------------------
# 1. Answer Relevancy Comparison
# ---------------------------
plt.figure(figsize=(13, 4))
plt.plot(df2["user_input"], df2["answer_relevancy"], marker='o', label="Answer_relevancy",color="orange")
plt.title("Answer relevancy")
plt.xlabel("Question")
plt.ylabel("Accuracy")
plt.ylim(0, 1.1)
plt.xticks(rotation=15, ha='right', fontsize=12)  
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Images/v2/answer_relevancy.png", dpi=300)  
plt.show()

# ---------------------------
# 2. Context Precision Comparison
# ---------------------------
plt.figure(figsize=(13, 4))
plt.plot(df2["user_input"], df2["context_precision_no_ref"], marker='o', label="Context_precision",color="orange")
plt.title("Context precision")
plt.xlabel("Question")
plt.ylabel("Accuracy")
plt.ylim(0, 1.1)
plt.xticks(rotation=15, ha='right', fontsize=12)  
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Images/v2/context_precision.png", dpi=300)  
plt.show()

# ---------------------------
# 3. Faithfulness Comparison
# ---------------------------
plt.figure(figsize=(13, 4))
plt.plot(df2["user_input"], df2["faithfulness"], marker='s', label="Faithfulness",color="orange")
plt.title("Faithfulness ")
plt.xlabel("Question")
plt.ylabel("Accuracy")
plt.ylim(0, 1.1)
plt.xticks(rotation=15, ha='right', fontsize=12)  
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Images/v2/faithfulness.png", dpi=300)  
plt.show()
```
