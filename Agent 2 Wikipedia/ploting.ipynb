```python
#import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_json("Results/tools_v1.json")
df2 = pd.read_json("Results/tools_v2.json")
```

```python
import matplotlib.pyplot as plt

# ---------------------------
# 1. Tool Accuracy Comparison
# ---------------------------
plt.figure(figsize=(12, 4))
plt.plot(df["question"], df["ragas_tool_accuracy"], marker='o', label="RAGAS Tool Accuracy",color="blue")
plt.title("RAGAS Tool Accuracy")
plt.xlabel("Question")
plt.ylabel("Accuracy")
plt.ylim(0, 1.1)
plt.xticks(rotation=15, ha='right', fontsize=12)  
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Images/v1/ragas_tool_accuracy.png", dpi=300)  
plt.show()

# -----------------------
# 2. Number of Reasoning Steps
# -----------------------
plt.figure(figsize=(12, 4))
plt.bar(df["question"], df["num_steps"], color="blue")
plt.title("Number of Reasoning Steps per Question")
plt.xlabel("Question")
plt.ylabel("Steps")
plt.xticks(rotation=15, ha='right', fontsize=12)  
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("Images/v1/num_steps.png", dpi=300)  

plt.show()

import matplotlib.pyplot as plt

# ---------------------------
# 1. Tool Accuracy Comparison
# ---------------------------
plt.figure(figsize=(12, 4))
plt.plot(df2["question"], df2["ragas_tool_accuracy"], marker='o', label="RAGAS Tool Accuracy",color="orange")
plt.title("RAGAS Tool Accuracy")
plt.xlabel("Question")
plt.ylabel("Accuracy")
plt.ylim(0, 1.1)
plt.xticks(rotation=15, ha='right', fontsize=12)  
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Images/v2/ragas_tool_accuracy.png", dpi=300)  
plt.show()

# -----------------------
# 2. Number of Reasoning Steps
# -----------------------
plt.figure(figsize=(12, 4))
plt.bar(df2["question"], df2["num_steps"], color="orange")
plt.title("Number of Reasoning Steps per Question")
plt.xlabel("Question")
plt.ylabel("Steps")
plt.xticks(rotation=15, ha='right', fontsize=12)  
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("Images/v2/num_steps.png", dpi=300)  

plt.show()
```
