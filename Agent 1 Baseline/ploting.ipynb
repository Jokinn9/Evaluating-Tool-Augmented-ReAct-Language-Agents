```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_json("Results/results_v1.json")
df2=pd.read_json("Results/results_v2.json")
df.head(2)
```

### Individually

```python
# ---------------------------
# 1. Tool Accuracy Comparison
# ---------------------------
plt.figure(figsize=(15, 5))
plt.plot(df["question"], df["tool_accuracy"], marker='o', label="Tool Accuracy")
plt.plot(df["question"], df["ragas_tool_accuracy"], marker='x', label="RAGAS Tool Accuracy")
plt.title("Tool Accuracy vs RAGAS Tool Accuracy")
plt.xlabel("Question")
plt.ylabel("Accuracy")
plt.ylim(0, 1.1)
plt.xticks(rotation=25, ha='right')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Images/v1/Tool.png", dpi=300)  
plt.show()

# ---------------------------------------
# 2. Output and Value Correct Flags (bar)
# ---------------------------------------
plt.figure(figsize=(15, 5))
output_flags = df["output_contains_expected"].astype(int)
value_flags = df["value_correct"].astype(int)

x = range(len(df))
bar_width = 0.35

plt.bar(x, output_flags, width=bar_width, label='Output Contains Expected', align='center')
plt.bar([p + bar_width for p in x], value_flags, width=bar_width, label='Value Correct', align='center')

plt.xticks([p + bar_width/2 for p in x], df["question"], rotation=25, ha='right')
plt.ylim(0, 1.1)
plt.ylabel("Binary Flag")
plt.title("Output & Value Correctness")
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("Images/v1/Value.png", dpi=300)
plt.show()

# -----------------------
# 3. Number of steps
# -----------------------
plt.figure(figsize=(15, 5))
plt.bar(df["question"], df["num_steps"], color="skyblue")
plt.title("Number of Reasoning Steps per Question")
plt.xlabel("Question")
plt.ylabel("Steps")
plt.xticks(rotation=25, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("Images/v1/Steps.png", dpi=300)
plt.show()
```
